## Machine Learning

* ML is programming computer to optimize a performance criterion using example data or past experience. ML can automatically detect patters in data, and then to use the uncovered patters to predict future data or other outcomes of interest.

* ML is based on statistics is bassically attempting to find the relationship between input and output variables.

* ML is the most popular technique of creating predictive and prescriptive model.

* ML is closely associated with Statistics, AI and Data Mining. 


* ML Vs. Statistics
    * Traditional Statistics focuses on provable results with mathematical assumptions, and care less about computation.
    * Statistics: A useful tool for ML.

* ML vs. AI
    * Machine LEarning is one possible route to realize AI.

* ML vs. Data mining
    * Traditional DM focuses on provable results with math assumptions along with efficiente computation in large dataset. 
    * It is difficult to distinguish ML and DM in reality


So for example when a real estate agent want to calculate a price of a property (Y), it used different input variables, such as area covered (x1), number of bedrooms(x2) and so on(x3...). With this he can calculate f(x). 
* There are two major reasons to why estimate f(x), which are:
    * Prediction: when the values of input variables is available and output variable is to be predicted. We are only interested in the value of Y, not in the relationship of Y with other variables.
    * Inference: When the relationship between input and output variable is important. We want to establish how output variable varies with change in each predictor variable.

When choosing a ML Model, we have to conider if we want to **predict** or **infer**.
So for **prediction**, *accuracy* of predicted function is the most important and for **inference**, *interpretability* of predicted fucntion is most important.
For example, **linear regression** is simple to interpret but may not give very accurate predicted values of Y whereas highly **non-linear models** may be predicting very accurately but hte relationship may be very difficult to interpret.

After this, we need to specify the type of **learning method**
Which can be a **parametric approach**, we assume the funcitonal form of the relationship between predictor and predicted variable. Y = a0 + a1 * x1 + ... + an*xn
in **non-parametric approach**, we don't assume any functional form for the relationship. Instead **f** is estimated by getting as close to the training points.


* Parametric approach:
    * Usually more interpretable.
    * May not be as accurate.
    * Preferable if inference is the reason estimating f(x).

* Non-parametric approach:
    * Less interpretable.
    * Potentially more accurate.
    * Needs large amount of data to train.
    * Preferable if prediction is the priority.


## Types of Learning:
* Supervised Learning:
    * Supervised learning is where you have input variables (x) and an output variable (Y) and you use an algorithm to learn the mapping function from the input to the output.
    * The goal is to approximate the mapping function so well that when you have new input data(x) that you can predict the output variables (Y) for that data.
    * SVM, K-NN, ANN and regression are examples of supervised learning.

* Unsupervised Lerning:
    * Unsupervised learning is where you only have input data (x) and no corresponding output variables.
    * The goal for unsupervised learning is to model the underlying structure or distribuition in the data in order to learn more about the data.
    * K-means, Hierachical clustering, Hidden Markov Models are examples of supervised learning.


## Building ML Model
1. Problem Formulation
    * Convert your bussiness problem into a Statistical problem.
    * Clearly define the dependent and independet variables.
    * Identify whether you want to predict or infer.
2. Data Tidying
    * Transform collected data into useable data table format.
3. Pre-processing
    * Filter data.
    * Aggregate values.
    * Missing value treatment.
    * Outlier treatment.
    * Variable transformation.
    * Variable reduction.
4. Train-Test Split
    * **Training data:** is the information used to train an algorithm. Training data inclues both input data and the corresponding expected output. Based on this data, the algorithm can learn the relationship between input and output variables.
    * **Testing data:** includes only input data, not the corresponding expected output. The testing data is used to asses the accuracy of model created or the predictor function created using the training data.
        * There shouldn't be any overlap between the two. 
        * Usually, 70-80% of the available data is used as training data and 20-303% as testing data. 
5. Model Training
    * y = f(x), where **y** is output, **f** is function and **x** is input variables.
6. Validation and Model accuracy
    * **Sample Error:** Error resulted from applying your prediction algorithm to the data set you built it with. 
    * **Out of Sample Error:** Error resulted from applying your prediction algorithm to a new data set.
7. Prediction
    * Setup a pipeline to use your model in real life scenario.
    * Improve by monitoring your model over time.
    * Try to automate.
